```python
import os
import shutil
from concurrent.futures import ThreadPoolExecutor

# --- PART 1: REBUILD DATASET (CORRECTED PATHS) ---
BASE = "/kaggle/input/ip02-dataset"
OUT = "/kaggle/working/ready"

# The 15 Target Classes
selected = {
    0: "rice leaf roller", 1: "rice leaf caterpillar", 3: "asiatic rice borer",
    4: "yellow rice borer", 5: "rice gall midge", 7: "brown plant hopper",
    8: "white backed plant hopper", 23: "army worm", 24: "aphids",
    27: "english grain aphid", 28: "green bug", 29: "bird cherry-oataphid",
    32: "longlegged spider mite", 33: "wheat phloeothrips", 34: "wheat sawfly"
}

# Clear previous failed attempts
if os.path.exists(OUT): shutil.rmtree(OUT)

# Create folders
for split in ["train", "val", "test"]:
    for cls in selected.values():
        os.makedirs(f"{OUT}/{split}/{cls}", exist_ok=True)

# Prepare file list
tasks = []
print("Rebuilding dataset with CORRECTED paths...")
for split in ["train", "val", "test"]:
    txt_path = f"{BASE}/{split}.txt"
    if not os.path.exists(txt_path): continue
    
    with open(txt_path) as f:
        lines = f.readlines()

    for line in lines:
        parts = line.strip().split()
        if len(parts) < 2: continue
        fname, lab = parts[0], int(parts[1])

        if lab in selected:
            # FIX: Use 'lab' directly (0-based), not 'lab+1'
            src = f"{BASE}/classification/{split}/{lab}/{fname}"
            dst = f"{OUT}/{split}/{selected[lab]}/{fname}"
            tasks.append((src, dst))

# Multi-threaded copy
def copy_file(t):
    if os.path.exists(t[0]): shutil.copy(t[0], t[1])

print(f"Copying {len(tasks)} images...")
with ThreadPoolExecutor(max_workers=8) as ex:
    list(ex.map(copy_file, tasks))

print("Dataset ready.")
```

```python
import os
from PIL import Image

root = "/kaggle/working/ready"

bad_images = []

for split in ["train", "val", "test"]:
    split_path = os.path.join(root, split)
    for cls in os.listdir(split_path):
        cls_path = os.path.join(split_path, cls)
        for img in os.listdir(cls_path):
            path = os.path.join(cls_path, img)
            try:
                Image.open(path).convert("RGB")
            except:
                bad_images.append(path)

print("Total corrupted images found:", len(bad_images))
bad_images[:20]
```

```python
import os
import cv2

root = "/kaggle/working/ready"

cv2_bad = []

for split in ["train", "val", "test"]:
    split_path = os.path.join(root, split)
    for cls in os.listdir(split_path):
        cls_path = os.path.join(split_path, cls)
        for img in os.listdir(cls_path):
            path = os.path.join(cls_path, img)
            im = cv2.imread(path)
            if im is None:
                cv2_bad.append(path)

print("OpenCV unreadable images:", len(cv2_bad))
cv2_bad[:20]
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n-cls.pt")

model.train(
    data="/kaggle/working/ready",
    epochs=30,
    imgsz=224,
    batch=32,
    amp=False,
    workers=0,
    device=0
)
```
